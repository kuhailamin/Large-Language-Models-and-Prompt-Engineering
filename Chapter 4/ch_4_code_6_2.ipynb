```python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT tokenizer and model
MODEL_NAME = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)
model = BertModel.from_pretrained(MODEL_NAME)

# Example sentences
sentences = ["BERT is amazing!", "Transformers are the future of NLP."]

# Tokenize and encode input text
inputs = tokenizer(sentences, return_tensors="pt", padding=True, truncation=True, max_length=512)

# Forward pass to extract embeddings
with torch.no_grad():
    outputs = model(**inputs)

# Extract last hidden states (token embeddings)
last_hidden_states = outputs.last_hidden_state  # Shape: (batch_size, sequence_length, hidden_size)

# Compute sentence embedding by averaging token embeddings (excluding padding tokens)
sentence_embeddings = last_hidden_states.mean(dim=1)  # Shape: (batch_size, hidden_size)

# Print embeddings
print("Sentence Embeddings:", sentence_embeddings)
```
