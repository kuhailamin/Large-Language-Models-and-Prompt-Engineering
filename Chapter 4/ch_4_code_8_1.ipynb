```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline

# Load a pre-trained DistilBERT model fine-tuned on the SQuAD dataset
MODEL_NAME = "distilbert-base-cased-distilled-squad"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForQuestionAnswering.from_pretrained(MODEL_NAME)

# Initialize QA pipeline
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)

# Provide context for the question
context = """
Albert Einstein was a theoretical physicist who developed the theory of relativity,
one of the two pillars of modern physics (alongside quantum mechanics).
His work is also known for its influence on the philosophy of science.
"""

# Define a question related to the context
question = "Who developed the theory of relativity?"

# Use the pipeline to extract the answer
answer = qa_pipeline(question=question, context=context)

# Display the result
print(f"Question: {question}")
print(f"Answer: {answer['answer']} (Confidence: {answer['score']:.4f})")
```
