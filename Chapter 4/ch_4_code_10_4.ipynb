```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load GPT-2
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Prompt for children's story
prompt = "Toby the curious turtle loved to explore the sea. One day, he found a glowing shell near a coral reef."

# Tokenize and generate
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
output = model.generate(input_ids, max_length=100, do_sample=True, temperature=0.9)

# Show output
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
